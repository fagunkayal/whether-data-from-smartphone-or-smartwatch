```python
def class_likelihood_params(class_training_complete_data):
    
    class_likelihood_params_dict = dict()
    
    for cols in class_training_complete_data.columns[2:4]:
        
        empty_dict = dict()
        
        if class_training_complete_data[cols].dtype == "int64":
            
            for value in complete_data[cols].unique():
                
                if value not in class_training_complete_data[cols].unique():
                    
                    empty_dict[value] = 1/(len(class_training_complete_data)+len(complete_data[cols].unique()))
                    
                else:
                        
                    empty_dict[value] = len(class_training_complete_data[class_training_complete_data[cols] == value])/len(class_training_complete_data)
                        
        else:
                    
            empty_dict["class_normal_mean"] = np.mean(class_training_complete_data[cols])
            empty_dict["class_normal_std"] = np.std(class_training_complete_data[cols])
                    
            class_likelihood_params_dict[cols] = empty_dict
                    
    return class_likelihood_params_dict
```

```python
sp_likelihood_params_dict = class_likelihood_params(training_complete_data[training_complete_data['Target'] == 1])
sw_likelihood_params_dict = class_likelihood_params(training_complete_data[training_complete_data['Target'] == 0])
```

```python
sp_likelihood_params_dict
sw_likelihood_params_dict
```

```python
import scipy.stats as s
```

```python
def class_conditioned_likelihood_per_row(params_dict, cv_complete_data_row):

    p_features_on_class = 1

    for cols in cv_complete_data.columns[2:4]:

        if cv_complete_data[cols].dtype == "int64":

            p_features_on_class = p_features_on_class * params_dict[cols][cv_complete_data_row[cols]]

        else:

            if cols == cv_complete_data.columns[4]:

                p_features_on_class = p_features_on_class * s.poisson.pmf(cv_complete_data_row[cols],params_dict[cols]["class_poisson_lambda"])

            else:

                p_features_on_class = p_features_on_class * s.norm.pdf(cv_complete_data_row[cols],
                                                               params_dict[cols]["class_normal_mean"],
                                                               params_dict[cols]["class_normal_std"])
        
        return p_features_on_class
```

```python
sp_prior = len(training_complete_data[training_complete_data['Target'] == 1])/len(training_complete_data)
sw_prior = len(training_complete_data[training_complete_data['Target'] == 0])/len(training_complete_data)
```

```python
def fetch_predictions(complete_data):

    predictions = list()

    for i in range(0,complete_data.shape[0]):

            sp_posterior_i = class_conditioned_likelihood_per_row(sp_likelihood_params_dict,complete_data.iloc[i,2:4]) * sp_prior

            sw_posterior_i = class_conditioned_likelihood_per_row(sw_likelihood_params_dict,complete_data.iloc[i,2:4]) * sw_prior

            predictions.append(sp_posterior_i > sw_posterior_i)

    return np.array(predictions)
```

```python
cv_predictions = fetch_predictions(cv_complete_data)
```

```python
cv_gt_labels = np.array(cv_complete_data.iloc[:,13])
```

```python
def fetch_performance_metrics(predictions,gt_labels):

    accuracy = np.count_nonzero(predictions == gt_labels)/gt_labels.shape[0]

    return {"Accuracy":accuracy}
```

```python
cv_performance_metrics = fetch_performance_metrics(cv_predictions,cv_gt_labels)
```
